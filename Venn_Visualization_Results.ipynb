# 4. Positive Tweets Classification results visualization (Venn Digram) through BOW, TF-IDF and H-TFIDF features

```python
from matplotlib_venn import venn3
from matplotlib import pyplot as plt
plt.figure(figsize=(10,10))

A001 = 79000 #BOW
A010 = 77452 #H-TFIDF
A011 = 1020   # BOW, H-TFIDF
A100 = 96522 #TF-IDF
A101 = 16302 # BOW, TF-IDF
A110 = 9519 #H-TFIDF, TF-IDF
A111 = 59094 # ALL
out = venn3(subsets = {'001': A001, '010': A010,'011': A011,'100': A100,'101': A101,'110': A110,'111': A111}, set_labels = ("TF-IDF", "H-TFIDF", "BOW"), alpha = 0.99)

out.get_label_by_id('001').set_text("Only BOW\n"+str(A001-A011-A101-A111)+"\n("+str(round((A001-A011-A101-A111)*100/(A001+A010+A100), 2))+"%)")
out.get_label_by_id('010').set_text("Only H-TFIDF\n"+str(A010-A011-A110-A111)+"\n("+str(round((A010-A011-A110-A111)*100/(A001+A010+A100), 2))+"%)")
out.get_label_by_id('100').set_text(" Only TF-IDF\n"+str(A100-A101-A110-A111)+"\n("+str(round((A100-A101-A110-A111)*100/(A001+A010+A100), 2))+"%)")
out.get_label_by_id('011').set_text("Only\n BOW & \nH-TFIDF\n"+str(A011)+"\n("+str(round((A011)*100/(A001+A010+A100), 2))+"%)")
out.get_label_by_id('101').set_text("Only BOW\n & TF-IDF\n"+str(A101)+"\n("+str(round((A101)*100/(A001+A010+A100), 2))+"%)")
out.get_label_by_id('110').set_text("Only H-TFIDF & \nTF-IDF\n"+str(A110)+"\n("+str(round((A110)*100/(A001+A010+A100), 2))+"%)")
out.get_label_by_id('111').set_text("COMMON\n"+str(A111)+"\n("+str(round((A111)*100/(A001+A010+A100), 2))+"%)")

out.get_patch_by_id('101').set_color('#ff66cc')
out.get_patch_by_id('010').set_color('#ff3333')
out.get_patch_by_id('100').set_color('#339966')
out.get_patch_by_id('001').set_color('#3399ff')
for text in out.set_labels:
    text.set_fontsize(18)
for text in out.subset_labels:
    text.set_fontsize(14)
    text.set_fontstyle("italic")
    text.set_color("W")
```
